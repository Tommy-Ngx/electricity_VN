#### DOMESTIC CITY TOTAL TRAFFIC ( FOR ALL CITIES )

```python
dom_city1 = dom.drop('CITY2',1)
dom_city1 = pd.DataFrame(dom_city1.groupby(['DATE','CITY1'])[['DATE','CITY1','TOTAL_PASSENGERS']].sum())
dom_city1.reset_index(inplace=True)
dom_city1.rename(columns={'CITY1':'CITY'},inplace=True)
dom_city1.head(15)
```

```python
dom_city2 = dom.drop('CITY1',1)
dom_city2 = pd.DataFrame(dom_city2.groupby(['DATE','CITY2'])[['DATE','CITY2','TOTAL_PASSENGERS']].sum())
dom_city2.reset_index(inplace=True)
dom_city2.rename(columns={'CITY2':'CITY'},inplace=True)
dom_city2.head(10)
```

```python
# Outer Join 2 tables 
dom_city = pd.merge(dom_city1,dom_city2, how='outer' , on=['DATE','CITY'])

# Fill NAN with 0
dom_city.fillna(0,inplace=True)

# Add two columns 
dom_city['TOTAL_PASSENGERS'] = dom_city['TOTAL_PASSENGERS_x'] + dom_city['TOTAL_PASSENGERS_y']
dom_city['TOTAL_PASSENGERS'] = dom_city['TOTAL_PASSENGERS'].astype(int)

# Extract columns 
dom_city = dom_city[['DATE','CITY','TOTAL_PASSENGERS']]
```

## AIRLINE WISE INTERNATIONAL TRAFFIC 

```python
airline = pd.read_excel('../DATASET/Updated_Dataset/Airlinewise Monthly International Air Traffic To And From The Indian Territory Excel.xlsx')
airline.head()
```

Process Date Column 

```python
airline['MONTH'] = airline['MONTH'].map({'JAN':1,'FEB':2, 'MAR':3, 'APR':4,'MAY':5,'JUN':6, 'JUL':7,\
                               'AUG':8,'SEP':9, 'OCT':10, 'NOV':11, 'DEC':12})

airline['DATE'] = pd.to_datetime(airline.YEAR.astype(str) + '-' + airline.MONTH.astype(str))

airline['DATE'] =  airline['DATE'] + pd.offsets.MonthEnd()
```

```python
airline = airline[['DATE','YEAR', 'MONTH', 'QUARTER', 'AIRLINE NAME', 'CARRIER TYPE',
       'PASSENGERS TO INDIA', 'PASSENGERS FROM INDIA', 'FREIGHT TO INDIA',
       'FREIGHT FROM INDIA']] 

airline.columns = airline.columns.str.replace(' ','_')

airline['TOTAL_PASSENGERS'] = airline['PASSENGERS_TO_INDIA'] + airline['PASSENGERS_FROM_INDIA']
airline['TOTAL_FREIGHT'] = airline['FREIGHT_TO_INDIA'] + airline['FREIGHT_FROM_INDIA']

airline.head()
```

SAVE Data

```python
airline.to_excel('Airline_Traffic_INTERNATIONAL.xlsx',index=False)
```

## COUNTRY WISE INTERNATIONAL TRAFFIC

```python
country = pd.read_excel('../DATASET/Updated_Dataset/Countrywise Quarterly International Air Traffic To And From The Indian Territory Excel.xlsx')
country.head()
```

```python
country['DATE'] = pd.to_datetime(country.YEAR.astype(str) + '-' + country.QUARTER.astype(str))

country['DATE'] =  country['DATE'] + pd.offsets.QuarterEnd(0)
```

```python
country = country[['DATE','YEAR', 'QUARTER', 'COUNTRY NAME', 
       'PASSENGERS TO INDIA', 'PASSENGERS FROM INDIA', 'FREIGHT TO INDIA',
       'FREIGHT FROM INDIA']] 

country.columns = country.columns.str.replace(' ','_')

country['TOTAL_PASSENGERS'] = country['PASSENGERS_TO_INDIA'] + country['PASSENGERS_FROM_INDIA']
country['TOTAL_FREIGHT'] = country['FREIGHT_TO_INDIA'] + country['FREIGHT_FROM_INDIA']

country.head()
```

```python
country['COUNTRY_NAME'] = country['COUNTRY_NAME'].map(\
                                                      lambda x : 'UNITED ARAB EMIRATES' if x=='UNITED ARAB EMIRATES*' \
                                                      else x)
```

SAVE Data 

```python
country.to_excel('Country_Traffic_INTERNATIONAL.xlsx',index=False)
```
